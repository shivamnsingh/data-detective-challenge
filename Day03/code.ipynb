```python
drop_cols = [
    "Unnamed: 0", "Unnamed: 6", "id", "object_id", "name",   # junk IDs, no signal
    "zip_code", "latitude", "longitude", "city", "state_code.1",  # redundant with is_CA/is_NY/etc dummies
    "closed_at",          # LEAKAGE — only exists when company already closed
    "category_code",      # redundant — same info as is_software/is_web/etc dummies
    "founded_at", "first_funding_at", "last_funding_at"  # redundant with age_first/age_last_funding_year
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])
```

```python
df['status']
```

```python
df["target"] = df["status"].apply(lambda x: 1 if str(x).lower() != "closed" else 0)
df = df.drop(columns=["status"])
```

```python
df["has_milestone"] = df["age_first_milestone_year"].notnull().astype(int)
df["age_first_milestone_year"] = df["age_first_milestone_year"].fillna(0)
df["age_last_milestone_year"] = df["age_last_milestone_year"].fillna(0)
```

```python
#LOG-TRANSFORM funding
df["funding_total_usd_log"] = np.log1p(df["funding_total_usd"])
```

```python
df = df.fillna(0)
```
